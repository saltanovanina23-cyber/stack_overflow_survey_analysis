```
python_mask = survey['LanguageHaveWorkedWith'].str.contains('Python',na=False)
df_py = survey[python_mask].copy()
```

```
df_py= df_py.dropna(subset=['ConvertedCompYearly','Country'])
```

```
py_prog_by_country = (df_py.groupby('Country')['ConvertedCompYearly']
    .agg(mean_compensation='mean', median_compensation='median')
    .sort_values('median_compensation', ascending=False))
```

```
py_prog_by_country.head(10)
```
